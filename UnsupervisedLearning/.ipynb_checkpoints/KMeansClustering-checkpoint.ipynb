### Importing libraries

```python
%matplotlib inline
from sklearn.datasets import make_blobs
import numpy as np
import matplotlib.pyplot as plt
```

### Creating a dataset for clustering

```python
# Create blobs
data = make_blobs(n_samples=200, n_features=2, centers=4, cluster_std=1.5, random_state=50)
# Create np array for data points
points = data[0]
# Visualize dataset in a scatter plot
plt.scatter(data[0][:,0], data[0][:,1], c=data[1], cmap='viridis')
plt.xlim(-15,15)
plt.ylim(-15,15)
```

## Description of K-means
**Algorithm** - The original k-means algorithm proceeds by randomly choosing a location for K centers in the dataset (random seeding). It assigns datapoints to the nearest centers (based on Euclidean distance). Then it recomputes the centers to be at the centroid of the data points assigned to it. After this first seed, and assignment, in all the next iterations, it reassigns data points to the centers nearest to it and recomputes the centroids. Once the location of the centroids do not change more than a threshold value, the algorithm is said to be converged.  

**Main assumption** - Reaching the best distribution of centroids requires the minimization of sum of squared distances or minimizing the variance of each cluster.  

**Advantages** - It is simple and fast.

**Limitations** - This is a simple, and powerful algorithm but it suffers from several limitations:
1. Provides no information about the right K to choose for a dataset
2. If the seeds are initialized badly, the centroids may be stuck in local minima
3. It always gives an answer whether or not it is meaningful to your dataset
4. Scaling input data can completely change the results
5. If clusters have different variances, kmeans will split them incorrectly

### Implementing randomly-seeded K-means

```python
from sklearn.cluster import KMeans
def plot_kmeans(y_km):
    plt.scatter(points[y_km ==0,0], points[y_km == 0,1], s=100, c='red')
    plt.scatter(points[y_km ==1,0], points[y_km == 1,1], s=100, c='black')
    plt.scatter(points[y_km ==2,0], points[y_km == 2,1], s=100, c='blue')
    plt.scatter(points[y_km ==3,0], points[y_km == 3,1], s=100, c='cyan')
    plt.xlim(-15,15)
    plt.ylim(-15,15)
```

```python
kmeans1 = KMeans(init='random',n_clusters=4)
kmeans1.fit(points)
y_km1 = kmeans1.fit_predict(points)
plot_kmeans(y_km1)
```

### Implementing Kmeans++ (starting with a smart seeding of centroids)
**Algorithm** - Since regular kmeans start at an interesting

```python
kmeans2 = KMeans(init='k-means++',n_clusters=4)
kmeans2.fit(points)
y_km2 = kmeans2.fit_predict(points)
plot_kmeans(y_km2)
```

**Conclusion** - For this simple dataset where clusterd were fairly separate from each other and had the same variance/size, we could predict well using both Kmeans algorithms.

## Using non-standard data

```python
# Create blobs
data = make_blobs(n_samples=[30,50], n_features=2, centers=[[-5,1],[2,2]], cluster_std=[0.5,4], random_state=50)
# Create np array for data points
points = data[0]
# Visualize dataset in a scatter plot
plt.scatter(data[0][:,0], data[0][:,1], c=data[1], cmap='viridis')
plt.xlim(-15,15)
plt.ylim(-15,15)
```

```python
kmeans1 = KMeans(init='random',n_clusters=2,n_init=1)
kmeans1.fit(points)
y_km1 = kmeans1.fit_predict(points)
plot_kmeans(y_km1)
```

```python
kmeans2 = KMeans(init='k-means++',n_clusters=2)
kmeans2.fit(points)
y_km2 = kmeans2.fit_predict(points)
plot_kmeans(y_km2)
```

**Conclusion** - Using kmeans, even the improved kmeans++, we see that a system with two centers - but varying sizes of clusters, cannot be characterized well with this method.
